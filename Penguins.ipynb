# `Data Cleaning and Preprocessing`
* **_Checking for NaN values_**

```python
df.isnull().sum()
```

* **_Filling NaN Values_**

```python
df['bill_length_mm']=df['bill_length_mm'].fillna(df.bill_length_mm.mean())
```

```python
df['bill_depth_mm']=df['bill_depth_mm'].fillna(df.bill_depth_mm.mean())
```

```python
df['flipper_length_mm']=df['flipper_length_mm'].fillna(df.flipper_length_mm.mean())
```

```python
df['body_mass_g']=df['body_mass_g'].fillna(df.body_mass_g.mean())
```

```python
df['sex']=df['sex'].fillna(method='bfill')
```

* **_dataset after data preprocessing_**

```python
df
```

* ### Heatmap

#### Q. Demonstrate the top 10 values of bill length, bill depth, flipper length and body mass of penguins.
